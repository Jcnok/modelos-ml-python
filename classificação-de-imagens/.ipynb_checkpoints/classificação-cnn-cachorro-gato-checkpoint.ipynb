```python
# Bibliotecas de uso geral
import pandas as pd
import numpy as np
from skimage.io import imread
from skimage.util import crop
import matplotlib.pylab as plt
import os, os.path
```

```python
#
#
#  Função que carrega imagens e faz um "crop" pelo centro da imagem
#  utilizando skimage.crop
#
#  Entradas : path = caminho com as imagens
#             offsetx = saltos nas bordas na largura
#             offsety = saltos nas bordas na altura
#             x_size = tamanho do crop em x
#             y_size = tamnho do crop em y
#             formats = lista de formatos válidos
# 
#  Saída : img_list = lista de imagens
# 
#          cada imagem vai ter dimensão x_size x y_size 
#
def load_crop_images_by_center(path, 
                               offsetx = 0,
                               offsety = 0,
                               x_size=128, 
                               y_size=64, 
                               formats=[".jpg",".png"]):
    
    img_list = []
    
    for f in os.listdir(path):    
        ext = os.path.splitext(f)[1]    
        if ext.lower() not in formats:
            continue
    
        im = imread(os.path.join(path, f))
    
        # Largura e altura da imagem
        largura = im.shape[0]
        altura  = im.shape[1]
    
        # Imagem cropada a partir do início
        
        x_crop = (largura-x_size)/2 if largura%2 == 0 else (largura-x_size-1)/2
        y_crop = (altura-y_size)/2 if altura%2 == 0 else (altura-y_size-1)/2
         
        imcrop = crop(im, ((x_crop+offsetx, x_crop-offsetx), 
                           (y_crop+offsety, y_crop-offsety), 
                           (0,0)), copy=False)         
    
        img_list.append(imcrop)
        
    return img_list
```

```python
# Pasta das imagens positivas
path_people = 'INRIAPerson/96X160H96/Train/pos/'

# Pastas das imagens negativas
path_no_people_one = 'INRIAPerson/Train/neg/'
path_no_people_two = 'INRIAPerson/Test/neg/'
```

```python
#
# Vamos carregar algumas imagens de não pessoas (fundos)
#
image_list_no_people = load_crop_images_by_center(path_no_people_one)
image_list_no_people = image_list_no_people + load_crop_images_by_center(path_no_people_two)
 
#
# Vamos carregar algumas imagens de pessoas
#
image_list_people = load_crop_images_by_center(path_people, offsetx=16, offsety=16)
```

```python
#
# Vamos utilizar o HOG da skimage
#
from skimage.feature import hog
from skimage.color import rgb2gray

#
# Primeiro vamos colocar os conjuntos das imagens com o mesmo tamanho
# para nossa amostra ficar balanceada (temos mais amostras de pessoas
# que de não pessoas)
#

image_list_people_bb = image_list_people[0:len(image_list_no_people)]

#
# Calculando as características (HOG) e construindo nosso conjunto (X, y)
#
nbins = 9
block_size = 16
cells_per_block = 2

#
# Estruturas auxiliares
#
people      = []
not_people  = []
y_people    = []
y_no_people = []

#
# Amostras de pessoas
#
for i in range(len(image_list_people_bb)):
    people.append(hog(rgb2gray(image_list_people_bb[i]), 
                           orientations=nbins, 
                           block_norm='L2',
                           pixels_per_cell=(block_size, block_size), 
                           cells_per_block=(cells_per_block, cells_per_block), 
                           multichannel=False))
    y_people.append(1)
    
#
# Amostras de não pessoas
#
for i in range(len(image_list_no_people)):
    not_people.append(hog(rgb2gray(image_list_no_people[i]), 
                           orientations=nbins, 
                           block_norm='L2',
                           pixels_per_cell=(block_size, block_size), 
                           cells_per_block=(cells_per_block, cells_per_block), 
                           multichannel=False))
    y_no_people.append(0)
    
#
# Concatenando
#
X = np.r_[np.array(people), np.array(not_people)]
y = np.r_[np.array(y_people), np.array(y_no_people)]
```
